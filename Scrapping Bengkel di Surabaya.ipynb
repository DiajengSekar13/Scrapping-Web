### 1. Scrapping Bengkel Motor Surabaya (Service Segala Jenis Sepeda Motor) | P2rtech

tautan : https://zaubee.com/biz/bengkel-motor-surabaya-service-segala-jenis-sepeda-motor-p2rtech-psj453dc

```python
url = "https://nicelocal.id/surabaya/autoservice/bengkel_motor_surabaya_service_segala_jenis_sepeda_motor_p2rtech/"  # menentukan url untuk diamvil datanya
req = requests.get(url)         # mengirim GET ke URL dan menyimpan respon variabel
```

```python
soup = BeautifulSoup(req.text, "html.parser") #buat objc beautifulsoup dengan teks respon dan mengatur ke parser html
print(req.text)  # mencetak teks respon
```

```python
import pandas as pd

df = pd.DataFrame()  # Membuat data frame
for div in soup.find_all('div', class_='box-main'):
    title = div.find('span', itemprop='name').text.strip() if div.find('span', itemprop='name') else ''
    telepon = div.find('a', class_='tel-phone js-phone-number')['href'][4:] if div.find('a', class_='tel-phone js-phone-number') else ''
    alamat = div.find('address', itemprop='address').text.strip() if div.find('address', itemprop='address') else ''

    df = df.append({'nama bengkel': title,
                    'telepon': telepon,
                    'alamat': alamat},
                   ignore_index=True)  # Menambahkan data ke data frame
    print(div, '\n\n')
    # Memproses dan mencetak elemen yang diproses
```

```python
describe = df.describe()
print(describe)
```
