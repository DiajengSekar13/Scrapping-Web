```python
class HitungKata:
    def __init__(self, nama_file, encoding='utf-8'):
        self.nama_file = nama_file
        self.encoding = encoding

    def hitung(self):
        kata_count = {}
        with open(self.nama_file, 'r', encoding=self.encoding, errors='ignore') as file:
            for line in file:
                words = line.split()
                for word in words:
                    if word in kata_count:
                        kata_count[word] += 1
                    else:
                        kata_count[word] = 1
        return kata_count

# Contoh penggunaan
hitung_kata = HitungKata('berita-1')
kata_count = hitung_kata.hitung()
print(f"Jumlah kata dalam file {hitung_kata.nama_file}:")
for kata, jumlah in kata_count.items():
    print(f"{kata}: {jumlah}")
```
